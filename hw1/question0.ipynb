```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors

import sys
 
# the setrecursionlimit function is
# used to modify the default recursion
# limit set by python. Using this,
# we can increase the recursion limit
# to satisfy our needs
 
sys.setrecursionlimit(10**6)
```

```python
# Create data
MAX_XY = 101
data = np.zeros((MAX_XY, MAX_XY))



# Main DFS function
def dfs(v):
    # Create a set to store visited vertices
    visited = set()

    # Call the recursive helper function to print DFS traversal
    dfs_helper(v, visited)

# Recusive DFS helper function
def dfs_helper(v, visited):
    # Mark the current node as visited and print it
    visited.add(v)
#     print(v, end=" ")
    
    # Change it to either unblocked white (70%) or blocked black (30%)
    data[v] = np.random.binomial(1,0.3)
    
    neighbors = set();
    if v[0] > 0:
        neighbors.add((v[0]-1,v[1]))
    if v[0] < MAX_XY-1:
        neighbors.add((v[0]+1,v[1]))
    if v[1] > 0:
        neighbors.add((v[0],v[1]-1))
    if v[1] < MAX_XY-1:
        neighbors.add((v[0],v[1]+1))


    # Recurse for all the vertices adjacent to this vertex
    for neighbor in neighbors:
        if neighbor not in visited:
            dfs_helper(neighbor, visited)


print("The following is each position during the DFS traversal starting from (2,2):")
dfs((2,2))

# Create discrete colormap
cmap = colors.ListedColormap(['white', 'black'])

# Plot and show
fig, ax = plt.subplots()
ax.imshow(data, cmap=cmap)
plt.xticks([]) # remove the tick marks by setting to an empty list
plt.yticks([]) # remove the tick marks by setting to an empty list
plt.show()
```
